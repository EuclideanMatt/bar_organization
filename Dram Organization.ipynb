```python
def process_categories(row, categories):
    # Flatten the nested dictionary to get bottom-level key-value pairs
    flat_categories = {}
    reverse_mapping = {}

    for main_cat, sub_cats in categories.items():
        for sub_cat, value in sub_cats.items():
            if isinstance(value, list):
                for v in value:
                    flat_categories[v.lower().replace(" ", "")] = sub_cat
                    reverse_mapping[sub_cat] = main_cat
            else:
                flat_categories[value.lower().replace(" ", "")] = sub_cat
                reverse_mapping[sub_cat] = main_cat
    
    # Normalize and map 'cat_2' column based on 'item' and 'category'
    item_value = row['item'].lower().replace(" ", "") if row['item'] is not None else ''
    category_value = row['category'].lower().replace(" ", "") if row['category'] is not None else ''

    cat_2_value = None
    for key, value in flat_categories.items():
        if key in item_value or key in category_value:
            cat_2_value = value
            break
    
    # Map the 'type' column based on 'cat_2' value
    type_value = reverse_mapping.get(cat_2_value, None)

    return cat_2_value, type_value
```

```python
def update_tables_in_db(db_path, categories):
    # Connect to SQLite database
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get a list of all tables
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()

    for table_name in tables:
        table_name = table_name[0]

        # Load table into a DataFrame
        df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)

        # Apply the process_categories function to the DataFrame
        df[['cat_2', 'type']] = df.apply(lambda row: process_categories(row, categories), axis=1, result_type='expand')

        # Add 'draught' column
        df['draught'] = df.apply(
            lambda row: 1 if (row['category'] is not None and 'draught' in row['category'].lower()) or 
                              (row['item'] is not None and row['item'][0].isdigit()) else 0,
            axis=1
        )

        # Add 'bottle' column
        df['bottle'] = df.apply(
            lambda row: 1 if (
                (row['category'] is not None and any(term in row['category'].lower() for term in ['bottle', 'bottled'])) or
                (row['item'] is not None and any(term in row['item'].lower() for term in ['bottle', 'bottled']))
            ) else 0,
            axis=1
        )

        # Save the updated DataFrame back to the database
        df.to_sql(table_name, conn, if_exists='replace', index=False)
    
    # Close the connection
    conn.close()
```

```python
# updating all the 

db_path = r"C:\Users\Epicm\Desktop\data_bases\dbs\planes\dram_db.db"

update_tables_in_db(db_path, categories)
```
